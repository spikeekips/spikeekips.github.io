I will try to compare the size of serialized data between the several serializing method in Python, especialy `json` vs. `msgpack`, these are heavily in use in the most of python applications. I will also compare the data with compressing, `zlib` and `gzip`.

## Prerequisits

```python
%pylab inline
```

```python
def _ensure_python_package (n, name=None, ) :
    _name = name if name else n
    try :
        return __import__(_name, None, None, [], )
    except ImportError :
        !pip install $n
        return __import__(_name, None, None, [], )
```

```python
import pandas

def make_series (_max, func_data, func_dumps, ) :
    _l = list()
    _index = _max() if callable(_max) else range(1, _max, )
    for i in  _index :
        _l.append(len(func_dumps(func_data(i, ), ), ), )

    return pandas.Series(_l, index=_index, dtype=int, )
```

```python
loremipsum = _ensure_python_package('loremipsum', )

import string

def func_data (n, ) :
    s = str()
    _m = n * 30
    while len(s) < _m :
        s += ' '.join(loremipsum.get_paragraphs(1, ))
    
    return dict(v=s[:_m], )
```

```python
func_series = lambda : range(100, 1001, 10, )
```

```python
import gzip
import StringIO

def encode_zlib (s, ) :
    return s.encode('zlib', )

def encode_gzip (s, ) :
    _s = StringIO.StringIO()
    with gzip.GzipFile(mode='wb', fileobj=_s, ) as f :
        f.write(s, )

    return _s.getvalue()

def decode_gzip (s, ) :
    _s = StringIO.StringIO(s, )
    with gzip.GzipFile(mode='r', fileobj=_s, ) as f :
        return f.read()

func_compress = lambda f, z='zlib' : lambda x : (encode_zlib if z == 'zlib' else encode_gzip)(str(f(x)))
```

## message size

```python
series = dict()
```

### raw string

```python
series['raw'] = make_series(func_series, func_data, str, )
series['raw(z)'] = make_series(func_series, func_data, func_compress(str), ) # zlib compression
series['raw(g)'] = make_series(func_series, func_data, func_compress(str, 'gzip', ), ) # gzip compression
```

### `json`

```python
import json
series['json'] = make_series(func_series, func_data, json.dumps, )
series['json(z)'] = make_series(func_series, func_data, func_compress(json.dumps, ), )
series['json(g)'] = make_series(func_series, func_data, func_compress(json.dumps, 'gzip', ), )
```

### `bson`

```python
_ensure_python_package('pymongo', )
from bson import BSON
series['bson'] = make_series(func_series, func_data, BSON.encode, )
series['bson(z)'] = make_series(func_series, func_data, func_compress(BSON.encode, ), )
series['bson(g)'] = make_series(func_series, func_data, func_compress(BSON.encode, 'gzip', ), )
```

### `msgpack`

```python
msgpack = _ensure_python_package('msgpack-python', 'msgpack', )
series['msgpack'] = make_series(func_series, func_data, msgpack.packb, )
series['msgpack(z)'] = make_series(func_series, func_data, func_compress(msgpack.packb, ), )
series['msgpack(g)'] = make_series(func_series, func_data, func_compress(msgpack.packb, 'gzip', ), )
```

### Result

```python
from IPython.display import display
import pandas
df = pandas.DataFrame(series, )
display(df[:50])
```

```python
import matplotlib.pyplot as plt
figsize(20,6)
```

### Size Comparison By Method

```python
df.loc[:, ('bson', 'json', 'msgpack', 'raw', )].sub(df.loc[:, 'raw'], axis='index', ).div(df.loc[:, 'raw'], axis='index', ).plot()
```

### Size Comparison By Method And zlib Compressed

```python
df.loc[:,('bson(z)', 'json(z)', 'msgpack(z)', 'raw(z)', )].sub(df.loc[:,'raw(z)'],axis='index', ).div(df.loc[:,'raw(z)'],axis='index', ).plot()
```

## Compress Rate By Method

```python
ax = plt.subplot(111)

for i in ('bson', 'json', 'msgpack', 'raw', ) :
    ax.plot(df.index, df.loc[:,'%s(z)' % i].sub(df.loc[:,i],axis='index', ).div(df.loc[:,i],axis='index', ), label=i, )
    
ax.legend()
plt.show()
```

## Compression Method: gzip vs. zlib

```python
ax = plt.subplot(111)

for i in ('bson', 'json', 'msgpack', 'raw', ) :
    ax.plot(df.index, df.loc[:,'%s(z)' % i].sub(df.loc[:,'%s(g)' % i], axis='index', ).div(df.loc[:, '%s(g)' % i],axis='index', ), label='%s z/g' % i, )

ax.legend()
plt.show()
```
